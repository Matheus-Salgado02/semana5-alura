```python
def aprovado():
  nota = []
  media = 0
  for i in range(1,5):
    entrada = float(input(f"Informe a nota de numero {i} : "))
    nota.append(entrada)

  menor = min(nota)
  maior = max(nota)

  media = sum(nota)/len(nota)

  if media >= 6:
    estado = "aprovado"

  else:
    estado = "reprovado"

  return (menor,maior,media,estado)
```

```python
menor,maior,media,estado = aprovado()

print(f"O(a) estudante obteve uma média de {media}, com a sua maior nota de {maior} pontos e a menor nota de {menor} pontos e foi {estado}")
```
